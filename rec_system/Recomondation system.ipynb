```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd  # Make sure to import pandas

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')

# Define stop words
stop_words = set(stopwords.words('english'))

def process_text(text):
    if isinstance(text, list):  # Check if text is a list
        text = ' '.join(text)  # Join list items into a single string
    elif pd.isna(text):  # Check for NaN values
        return []  # Return an empty list for NaN
    text = text.casefold()  # Lowercase the text
    tokens = word_tokenize(text)  # Tokenize the text into words
    cleaned_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]  # Remove stop words
    return cleaned_tokens

# Apply the function to the relevant columns
dataset['product_title'] = dataset['product_title'].apply(lambda x: process_text(x))

product['product_title'] = product['product_title'].apply(lambda x: process_text(x))
product['category'] = product['category'].apply(lambda x: process_text(x))
product['brand_name'] = product['brand_name'].apply(lambda x: process_text(x))
product['product_color'] = product['product_color'].apply(lambda x: process_text(x))
product['availability_status'] = product['availability_status'].apply(lambda x: process_text(x))
product['manufacturer_name'] = product['manufacturer_name'].apply(lambda x: process_text(x))
```

```python
def recommend_similar_products(product_name, category, top_n):
    # Check if the product exists in the DataFrame
    product_indices = product[
        product['product_title'].str.contains(product_name, case=False, na=False) &
        product['category'].str.contains(category, case=False, na=False)
    ].index

    if len(product_indices) == 0:
        print(f"No product found matching '{product_name}' in the dataset.")
        return []

    # Use the first matching product for recommendations
    product_index = product_indices[0]

    # Get cosine similarity scores for the selected product
    distances = cosinesimilarity_matrix[product_index]
    print(distances)
    # Get most similar products excluding the input product itself
    similar_products_indices = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[0:top_n]

    # Compile recommended products into a list
    recommended_products = []
    print(f"\nProducts similar to '{product.iloc[product_index]['product_title']}':")
    for i in similar_products_indices:
        recommended_product = product.iloc[i[0]]
        recommended_products.append(recommended_product['product_title'])
        print(f"- {recommended_product['product_title']} (Similarity: {i[1]:.4f})")

    return recommended_products

# Usage example:
# Assuming `product` DataFrame and `cosinesimilarity_matrix` are already defined
recommended = recommend_similar_products("featherlite","shirt",5)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Convert the transactions into one-hot encoded format
Tencorder = TransactionEncoder()
Encoded_transactions = Tencorder.fit(transactions).transform(transactions)

# Each column corresponds to a unique product from the dataset. 
# Each row represents a single transaction.
# 1 indicates that the product is present in the corresponding transaction. 0 indicates that the product is absent from that transaction.
Encoded_Tdataframe = pd.DataFrame(Encoded_transactions, columns=Tencorder.columns_)

# Step 2: Apply FP-Growth algorithm to get frequent itemsets
# 'min_support' defines the minimum support threshold
frequent_itemsets = fpgrowth(Encoded_Tdataframe, min_support=0.01, use_colnames=True)

# Step 3: Generate association rules from the frequent itemsets
# 'metric' is set to 'confidence', and we filter rules with confidence >= 0.6
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Display results
print(frequent_itemsets)
rules.head()
```

```python
# Get top n rules related to items purchased using FP-Growth

def fpgrowth_recommendations(items, top_n):
    recommendations = {}
    for item in items:
        # Filter rules where the item is in the antecedents
        rule_subset = rules[rules['antecedents'].apply(lambda x: item in x)]
        # Increment the count for each consequent
        for consequent in rule_subset['consequents']:
            recommendations[consequent] = recommendations.get(consequent, 0) + 1

    # Sort by frequency and return the top_n recommendations
    return pd.Series(recommendations).sort_values(ascending=False).head(top_n)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Convert the transactions into one-hot encoded format
Tencorder = TransactionEncoder()
Encoded_transactions = Tencorder.fit(transactions).transform(transactions)

#Each column corresponds to a unique product from the dataset.   #Each row represents a single transaction.
#1 indicates that the product is present in the corresponding transaction. 0 indicates that the product is absent from that transaction.
Encoded_Tdataframe = pd.DataFrame(Encoded_transactions, columns=Tencorder.columns_)

# Step 2: Apply Apriori algorithm to get frequent itemsets
# 'min_support' defines the minimum support threshold
frequent_itemsets = apriori(Encoded_Tdataframe, min_support=0.01, use_colnames=True)

# Step 3: Generate association rules from the frequent itemsets
# 'metric' is set to 'confidence', and we filter rules with confidence >= 0.6
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Display results
print(frequent_itemsets)
rules.head()
```

```python
#get top n rules related to items purchased

def apriori_recommendations(items, top_n):
    recommendations = {}
    for item in items:
        rule_subset = rules[rules['antecedents'].apply(lambda x: item in x)]
        for consequent in rule_subset['consequents']:
            recommendations[consequent] = recommendations.get(consequent, 0) + 1

    return pd.Series(recommendations).sort_values(ascending=False).head(top_n)
```

```python
#user based filtering

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder

# Create a user-item matrix
user_item_matrix = dataset.pivot_table(index='user_id', columns='product_title', values='rating', aggfunc='mean', fill_value=0)

# Calculate cosine similarity between users
"""
ProdName  ProductA  ProductB  ProductC  ProductD
user_id
1              1        1        0        0
2              1        0        1        0
3              0        1        1        1

"""
user_similarity = cosine_similarity(user_item_matrix)

#abeling rows and columns with user IDs for easier interpretation.
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
"""
           1         2         3
1  1.000000  0.577350  0.000000
2  0.577350  1.000000  0.666667
3  0.000000  0.666667  1.000000

"""
print(user_similarity_df.shape)

# Display user similarity matrix
# print(user_similarity_df)

# Function to get recommendations for a user
#function that takes a user ID & give no of recomandatins
def get_user_recommendations(user_id, n_recommendations, diversify=True):
    # Get the most similar users to the given user
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:]  # Exclude self
    recommendations = []

    #print(f"Recommendations for User: {user_id}")
    #print(f"Number of Similar Users: {len(similar_users)}")
    #print(f"Similar Users: {similar_users.tolist()}")

    #Iterates through each similar user.
    #Retrieves items that the similar user has interacted with (purchased).
    #Adds these items to the recommendations list
    for similar_user in similar_users:
        recommended_items = user_item_matrix.loc[similar_user][user_item_matrix.loc[similar_user] > 0].index
        recommendations.extend(recommended_items)

    # Filter already purchased items
    #Removes items that the original user has already purchased from the recommendations
    recommendations = set(recommendations) - set(user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] > 0].index)
    return list(recommendations)[:n_recommendations]

# Example: Get recommendations for user_id A1007NP5HDQ8N7
print("user based recomendations for user A2X26X04JOOKOA :")
print(get_user_recommendations('A2X26X04JOOKOA',10))

print("\n user based recomendations for user A1007NP5HDQ8N7 :")
print(get_user_recommendations('A1007NP5HDQ8N7',10))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Create an Item-User Matrix
item_user_matrix = dataset.pivot_table(index='product_title', columns='user_id', values='rating', aggfunc='mean', fill_value=0)

# Step 2: Calculate Item Similarity
item_similarity = cosine_similarity(item_user_matrix)
item_similarity_df = pd.DataFrame(item_similarity, index=item_user_matrix.index, columns=item_user_matrix.index)

# Step 3: Item-Based Recommendations Function
def get_item_based_recommendations(user_id, n_recommendations):
    # Check if the user exists
    if user_id not in item_user_matrix.columns:
        print(f"No user found with ID '{user_id}'.")
        return []

    # Get items the user has interacted with
    user_items = item_user_matrix.loc[:, user_id]
    interacted_items = user_items[user_items > 0].index

    # Create a dictionary to store item similarity scores
    item_scores = {}

    for item in interacted_items:
        # Get similar items (exclude the item itself)
        similar_items = item_similarity_df[item].sort_values(ascending=False)[1:]

        for similar_item, score in similar_items.items():
            if similar_item not in interacted_items:  # Exclude items the user has already interacted with
                if similar_item in item_scores:
                    item_scores[similar_item] += score
                else:
                    item_scores[similar_item] = score

    # Sort items by similarity score
    recommended_items = sorted(item_scores.items(), key=lambda x: x[1], reverse=True)

    # Return the top N recommendations
    return [item[0] for item in recommended_items][:n_recommendations]

# Example: Get item-based recommendations for a user
print("\nItem-Based Recommendations for user 'A1007NP5HDQ8N7':")
print(get_item_based_recommendations('A1007NP5HDQ8N7',5))
print("\nItem-Based Recommendations for user 'A2X26X04JOOKOA':")
print(get_item_based_recommendations('A2X26X04JOOKOA',5))
print("\nItem-Based Recommendations for user 'A34F9C6WGMQTEF':")
print(get_item_based_recommendations('A34F9C6WGMQTEF',5))
```

```python
import re

def is_item_in_category(item, category):
    print(f"Searching for item: '{item}'")  # Debugging output

    # Check if item is in any product title
    item_in_titles = dataset['product_title'].str.contains(r'\b' + re.escape(item) + r'\b', case=False, na=False)

    if item_in_titles.any():
        print("Item found in titles.")  # Debugging output
        # If item found in titles, retrieve the corresponding categories
        matched_categories = dataset.loc[item_in_titles, 'category'].unique()
        matched_categories_lower = [cat.lower() for cat in matched_categories]  # Convert to lowercase
        print(f"Matched categories: {matched_categories}")  # Debugging output

        result = category.lower() in matched_categories_lower  # Case-insensitive match
        print(f"Category match result: {result}")  # Debugging output
        return result  # Return True if the category matches any found

    print("Item not found in titles.")  # Debugging output
    return False  # Return False if no match is found
```

```python
def hybrid_recommendation(user, product_title, category, no_of_recomendations, w_user_user=0.1, w_item_item=0.3, w_content=0.4,  w_rule=0.2):
    # Get recommendations from each filter
    user_user_score = get_user_recommendations(user,no_of_recomendations)
    item_item_score = get_item_based_recommendations(user,no_of_recomendations)

    # Call similar products and handle potential None return
    content_score = recommend_similar_products(product_title,category,top_n=5)
    if content_score is None:
        content_score = []  # Handle None case

    #apirori rules
    apirori_rule_score = fpgrowth_recommendations(content_score, top_n=5)

    # Convert lists to sets for set operations
    user_user_set = set(user_user_score)
    item_item_set = set(item_item_score)
    content_set = set(content_score)
    rules_set = set(apirori_rule_score)

    # Combine recommendations using weighted hybrid approach
    final_recommendations = []
    for item in user_user_set.union(item_item_set).union(content_set).union(rules_set):
        score = 0
        if item in user_user_set:
            score += w_user_user
        if item in item_item_set:
            score += w_item_item
        if item in content_set:
            score += w_content
        if item in rules_set:
            score += w_rule
        #final_recommendations.append((item, score))

        # You should check if the item belongs to the specified category here
        if is_item_in_category(item, category):  # You need to implement this function
          final_recommendations.append((item, score))

    # Sort recommendations by score and return top recommendations
    final_recommendations.sort(key=lambda x: x[1], reverse=True)
    return final_recommendations  # Return item and score

# Example usage
user = 'A2VGOW1GQ7IXIZ'
product_title = 'featherlite'
category = 'shirt'
no_of_recs = 5
recommendations = hybrid_recommendation(user, product_title, category, no_of_recs)

# Format output as bullet points
print(f'\n\nHybrid recommendations for user {user} and product "{product_title} and category {category}":')
for item, score in recommendations:
    print(f"- {item} (Score: {score:.2f})")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming we have ground truth data and predictions
# Ground truth: list of items that the user actually interacted with
# Predicted: list of items recommended by the system

def evaluate_recommendations(ground_truth_list, recommended):
    # Convert both ground truth and predictions to sets for evaluation
    ground_truth_set = set(ground_truth_list)
    predicted_set = set(recommended)

    # Intersection gives true positives
    true_positives = ground_truth_set & predicted_set

    precision = len(true_positives) / len(predicted_set) if len(predicted_set) > 0 else 0
    recall = len(true_positives) / len(ground_truth_set) if len(ground_truth_set) > 0 else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1

precision, recall, f1 = evaluate_recommendations(ground_truth_list, recommended)
print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
```
